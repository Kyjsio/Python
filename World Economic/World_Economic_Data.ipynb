## Corruption

```python
t15_corr=df_corr.sort_values('annual_income',ascending=False).head(15)
```

```python
fig, corr=plt.subplots(figsize=(16, 6))

corr.set_ylabel('Annual Income', color='skyblue')
corr.bar(t15_corr['country'], t15_corr['annual_income'],color='skyblue')
corr.tick_params(axis='y', labelcolor='skyblue')
corr.set_xticks(range(len(t15_corr['country'])))
corr.set_xticklabels(t15_corr['country'],rotation=90)
corr.set_xlabel('Country')
corr.set_title('Richest country vs corruption')
inc=corr.twinx()

inc.set_ylabel('Corruption Index', color='blue')
inc.plot(t15_corr['country'], t15_corr['corruption_index'],color='blue',marker="o")
inc.tick_params(axis='y', labelcolor='blue')
plt.show()
```

```python
df_corr_v2=df_corr.copy()
df_corr_v2.loc[df_corr_v2['corruption_index'].between(0, 20, 'right'), 'corruption_index_group'] = '0-20'
df_corr_v2.loc[df_corr_v2['corruption_index'].between(20, 40, 'right'), 'corruption_index_group'] = '20-40'
df_corr_v2.loc[df_corr_v2['corruption_index'].between(40, 60, 'right'), 'corruption_index_group'] = '40-60'
df_corr_v2.loc[df_corr_v2['corruption_index'].between(60, 80, 'right'), 'corruption_index_group'] = '60-80'
df_corr_v2.loc[df_corr_v2['corruption_index'].between(80, 100, 'right'), 'corruption_index_group'] = '80-100'

sns.boxplot(data =df_corr_v2, x = 'corruption_index_group', y = 'annual_income')
plt.grid(True)
plt.xlabel('Corruption index')
plt.ylabel('Annual income')
plt.title('Group of Corruption')
```

The data analysis shows the country with the highest annual income have corruption index between 0-20 and median is a bit above a 60000 annual income.Interestigly looks group 80-100 ,where median income is close to 0 and this same time few countries is richest than some countries in group 60-80 with less corruption.

## Cost_of_living

```python
df_cost.head(10)
```

```python

df_cost_inc=df_cost.sort_values('monthly_income',ascending=False)[:30]

fig, pl1 = plt.subplots(figsize=(16, 6))

pl1.set_ylabel('Monthly Income', color='skyblue')
pl1.bar(df_cost_inc['country'], df_cost_inc['monthly_income'],color='skyblue')
pl1.tick_params(axis='y', labelcolor='skyblue')
pl1.set_title('Monthly income vs Cost index')
pl1.set_xticks(range(len(df_cost_inc['country'])))
pl1.set_xticklabels(df_cost_inc['country'], rotation=90)
pl1.set_xlabel('Country')
pl2 = pl1.twinx() 

pl2.set_ylabel('Cost Index', color='blue')
pl2.plot(df_cost_inc['country'], df_cost_inc['cost_index'], color='blue', marker='o')
pl2.tick_params(axis='y', labelcolor='blue')

plt.show()
```

## Combined data 

```python
all_df= df_corr.merge(df_unemployment, on = 'country', how = 'inner', sort = True).merge(df_richest, on = 'country', how = 'inner', sort = True).merge(df_cost, on = 'country', how = 'inner', sort = True)
all_df = all_df.sort_values(by =['annual_income','corruption_index','unemployment_rate','gdp_per_capita','purchasing_power_index'], ascending=False).reset_index(drop=True)
all_df.style.background_gradient()
```
